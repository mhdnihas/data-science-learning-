### Uniberivate analysis:

```python
categorical_col = []
for column in df.columns:
    if df[column].dtype == object and len(df[column].unique()) <= 50:
        categorical_col.append(column)
        print(f"{column} : {df[column].unique()}")
        print("====================================")
del categorical_col[0]
len(categorical_col)
```

```python
numerical_col = []#list for continous columns
for column in df.columns:#acessing columns from datasets
    
    if df[column].dtype == 'int64' and len(df[column].unique()) >= 10: 
       
        numerical_col.append(column)                           
numerical_col       
```

```python
plt.figure(figsize=(15,65))
plt.suptitle('uniberivate continouse values')
plotnum=0
for column in numerical_col:
    plt.suptitle('uniberivate continouse values')
    plotnum+=1
    plt.subplot(14,2,plotnum)
    sns.distplot(df[column])

    plotnum+=1
    plt.subplot(14,2,plotnum)
    sns.boxenplot(x=df[column])
    plt.xlabel(column,fontsize=20)
```

```python
plt.figure(figsize=(30,10))
plotnumber = 1

for column in categorical_col:
    ax = plt.subplot(2,4,plotnumber)
    sns.countplot(x=df[column],hue=df.Attrition)
    plt.xlabel(column,fontsize=20)
    plt.ylabel('Attrition',fontsize=20)
    
    plt.xticks(rotation=45)
    plotnumber+=1
```
